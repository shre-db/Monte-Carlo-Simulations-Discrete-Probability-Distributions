```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

## **Discrete Uniform Distribution**

The discrete uniform distribution is simply a probability distribution that places equal probabilities for all equal outcomes. One example is given by the probability of the outcomes of a die toss. The probability of each possible outcome for a fair, six-sided die is given by,

$P(X = x) = \frac{1}{6} \hspace{5mm} for \hspace{1mm} x=1,...,6$

```python
# Monte Carlo Simulation of Discrete Uniform Distribution

min_val, max_val, N = 1, 6, 10**4

# Run the simulation
mc_estimate = [np.random.randint(min_val, max_val+1) for i in range(N)]

# Average
print(f"Mean: {sum(mc_estimate)/N}")

# Plot
plt.hist(mc_estimate, bins=range(min_val, max_val + 2), align="left", density=True)
plt.title("Monte Carlo Simulation of Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

## **Bernoulli Distribution**

The Bernoulli distribution models a binary outcome, such as success/failure, yes/no, or heads/tails. It has a single parameter p, which is the probability of success.

$P(X=x) = p^x(1-p)^{1-x} \hspace{5mm} for \hspace{1mm}x \in \{0,1\}$

```python
# Monte Carlo Simulation of a Bernoulli Distribution

P = 0.3
N = 10**4

# Run the simulation
mc_estimate = [1 if np.random.random() <= P else 0 for trial in range(N)]

successes = mc_estimate.count(1)
failures = mc_estimate.count(0)

print(f"Mean: {sum(mc_estimate)/N}")
plt.bar(['Success', 'Failure'], [successes, failures], align='center')
plt.title("Monte Carlo Simulation of Bernoulli Distribution")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.show()
```

## **Binomial Distribution**

The Binomial distribution models the number of successes in a fixed number of independent Bernoulli trials. It has two parameters: $n$ (number of trials) and $p$ (probability of success in each trial).

$P(X=x) = \bigg ( \begin{matrix}n \\ k\end{matrix} \bigg) p^k (1-p)^{n-k} \hspace{5mm} for \hspace{1mm} k=0,1,2...,n.$

```python
# Monte Carlo Simulation of a Binomial Distribution
 
n = 10  # Number of trials
P = 0.3  # Probability of success for each trial
N = 10000  # Number of Simulations

samples = []

# Run Monte Carlo Simulation
for sim in range(N):
    successes = 0
    for trial in range(n):
        if np.random.random() <= P:
            successes += 1
    samples.append(successes)

print(f"Mean: {sum(samples)/N}")
plt.hist(samples, bins=range(n+2), align='left', density=True)
plt.title("Monte Carlo Simulation of Binomial Distribution")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

## **Poisson Distribution**

The Poisson distribution models the number of events occurring in a fixed interval of time or space. It has one parameter $\lambda$, which represents the average rate of occurrence.

$P(X=k) = \frac{\large e^{-\lambda}\lambda^k}{\large k!} \hspace{5mm} for \hspace{1mm} k=0,1,2,...$

```python
# Monte Carlo Simulation of a Poisson Distribution

lam = 5.5  # Average Rate
N = 10**5  # Number of Simulation

samples = []


# Run Monte Carlo simulation
for sim in range(N):
    k = np.random.poisson(lam=lam)
    samples.append(k)

print(f"Mean: {sum(samples)/N}")
plt.hist(samples, bins=range(max(samples)), align="left", density=True)
plt.title("Monte Carlo Simulation of Poisson Distribution")
plt.xlabel("Number of events")
plt.ylabel("Probability")
plt.show()
```

## **Geometric Distribution**

The Geometric Distribution models the number of trials needed to achieve the first success in a sequence of independent Bernoulli trials. It has one parameter $p$ (probability of success in each trial).

$P(X=k) = p(1-p)^{k-1} \hspace{5mm} for \hspace{1mm} k = 1,2,3,...$

```python
# Monte Carlo Simulation of Geometric Distribution

P = 0.3
N = 10**5

# Run the Monte Carlo Simulation
samples = []
for sim in range(N):
    trials_until_success = 1
    while np.random.random() > P:
        trials_until_success += 1
    samples.append(trials_until_success)

print(f"Mean: {np.mean(samples)}")
plt.hist(samples, bins=range(1, max(samples)), align="left", density=True)
plt.title("Monte Carlo Simulation of Geometric Distribution")
plt.xlabel("Number of trials until success")
plt.ylabel("Probability")
plt.show()
```

## **Negative Binomial Distribution**

The Negative Binomial Distribution models the number of trials needed to achieve a fixed number of successes in a sequence of independent Bernoulli trials.

It has two parameters: $r$ (number of successes) and $p$ (probability of success in each trial).

$P(X=k) = \bigg (\begin{matrix}k-1 \\ r-1\end{matrix} \bigg )p^r(1-p)^{k-r} \hspace{5mm} for \hspace{1mm} k=r, r+1, r+2,...$

```python
# Monte Carlo Simulation of a Negative Binomial Distribution

P = 0.3
r = 5
N = 10**5

samples = []

# Run the Simulation
for sim in range(N):
    successes = 0
    trials = 0
    while successes < r:
        prob = np.random.random()
        if prob <= P:
            successes += 1
        trials += 1
    samples.append(trials)

print(f"Mean: {np.mean(samples)}")
plt.hist(samples, bins=range(max(samples)), align='left', density=True)
plt.title("Monte Carlo Simulation of Negative Binomial Distribution")
plt.xlabel("Number of trials until 'r' successes.")
plt.ylabel('Probability')
plt.show()
```

## **Hypergeometric Distribution**

The hypergeometric distribution models the number of successes in a sample drawn without replacement from a finite population.

It has three parameters: 

$N$  population size,

$K$  number of successes in the population

$n$  sample size

$P(X=k) = \frac{\bigg( \begin{matrix} K \\ k\end{matrix}\bigg) \bigg( \begin{matrix} N-K \\ n-k\end{matrix}\bigg)}{\bigg( \begin{matrix}N \\ n\end{matrix} \bigg)} \hspace{5mm} for \hspace{1mm} k=max(0, n - (N-K)), ..., min(K, n)$

```python
# Monte Carlo Simulation of Hypergeometric Distribution

N = 100  # Population Size
K = 60  # Number of successes in the population
n = 10  # Sample Size
sims = 10**5  # Number of Simulations

samples = []

# Run the simulation
for sim in range(sims):
    # Create a population with K successes and N-K failures
    population = np.concatenate([np.ones(K), np.zeros(N - K)])

    # Randomly select a sample of size n without replacement
    sample = np.random.choice(a=population, size=n, replace=False)

    # Count the number of successes in the sample
    successes = np.sum(sample)
    samples.append(successes)

print(f"Mean: {sum(samples)/sims}")
plt.hist(samples, bins=range(n+1), align='left', density=True)
plt.title('Monte Carlo Simulation of Hypergeometric Distribution')
plt.xlabel('No of successes in the sample')
plt.ylabel('Probability')
plt.show()
```
